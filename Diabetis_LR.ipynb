```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets,linear_model
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

```python
diabetis =datasets.load_diabetes()
```

```python
print(diabetis.keys())
```

```python
print(diabetis.DESCR)
```

```python
import pandas as pd
df = pd.DataFrame(diabetis.data, columns=diabetis.feature_names)
df.head()
```

```python
diabetis_X=diabetis.data[:,np.newaxis,2]
```

```python
print(diabetis_X)
```

```python
diabetis_X_train=diabetis_X[:-30]
diabetis_X_test=diabetis_X[-20:]

diabetis_Y_train=diabetis.target[:-30]
diabetis_Y_test=diabetis.target[-20:]
```

```python
model=linear_model.LinearRegression()

#distinguising data by a line

model.fit(diabetis_X_train,diabetis_Y_train)
```

```python
#putting data under prediction

diabetis_Y_predict=model.predict(diabetis_X_test)
```

```python
#Checking error with mean square error

print("Mean square error: ", mean_squared_error(diabetis_Y_test,diabetis_Y_predict))
print("Weight: ",model.coef_)
print("Intercept: ",model.intercept_)
```

```python
#Comparison with scatter plot and line with single weight value
plt.scatter(diabetis_X_test,diabetis_Y_test)
plt.plot(diabetis_X_test,diabetis_Y_predict,c='orange')
plt.show()
```

```python
#Considering Multiple weights
diabetis_X=diabetis.data
diabetis_X_train=diabetis_X[:-30]
diabetis_X_test=diabetis_X[-20:]

diabetis_Y_train=diabetis.target[:-30]
diabetis_Y_test=diabetis.target[-20:]

model=linear_model.LinearRegression()
model.fit(diabetis_X_train,diabetis_Y_train)

diabetis_Y_predict=model.predict(diabetis_X_test)

print("Mean square error: ", mean_squared_error(diabetis_Y_test,diabetis_Y_predict))
print("Weight: ",model.coef_)
print("Intercept: ",model.intercept_)
```
